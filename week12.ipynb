1번문제 

```python
import csv

f = open("기공데.csv", "r")

rd = csv.reader(f)
next(rd)

maxdif = 0
difDay = ""
mintem = 100
minDay = ""
maxtem = 0
maxDay = ""
for ln in rd:
    Matem = float(ln[4])
    Mitem = float(ln[6])
    Diftem = float(ln[8])

    Day = ln[2]
    if(Diftem > maxdif):
        print("True")
        maxdif = Diftem
        difDay = Day
    if (mintem > Mitem):
        mintem = Mitem
        minDay = Day
    if (maxtem < Matem):
        mixtem = Matem
        maxDay = Day

print("일교차가 가장 큰 날")
print(difDay)
print("최저기온이 가장 낮은 날")
print(minDay)
print("최고기온이 가장 높은 날" )
print(maxDay)
```
